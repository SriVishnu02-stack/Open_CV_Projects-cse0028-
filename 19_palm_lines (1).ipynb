```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read image
image = cv2.imread("palm.jpg")

if image is None:
    print("Image not found!")
else:
    # Original Image
    plt.figure(figsize=(5,5))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")
    plt.show()

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    plt.figure(figsize=(5,5))
    plt.imshow(gray, cmap="gray")
    plt.title("Gray Image")
    plt.axis("off")
    plt.show()

    # Canny Edge Detection
    edges = cv2.Canny(gray, 50, 150)

    plt.figure(figsize=(5,5))
    plt.imshow(edges, cmap="gray")
    plt.title("Detected Palm Lines")
    plt.axis("off")
    plt.show()

    # Create a blank image for colored lines
    colored_lines = np.zeros_like(image)

    # Highlight detected edges in Yellow (BGR)
    colored_lines[edges == 255] = (0, 255, 255)

    plt.figure(figsize=(5,5))
    plt.imshow(cv2.cvtColor(colored_lines, cv2.COLOR_BGR2RGB))
    plt.title("Colored Palm Lines")
    plt.axis("off")
    plt.show()

    # Blend colored lines with original image
    result = cv2.addWeighted(image, 1.0, colored_lines, 0.8, 0)

    plt.figure(figsize=(6,6))
    plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    plt.title("Final Highlighted Palm")
    plt.axis("off")
    plt.show()
```
